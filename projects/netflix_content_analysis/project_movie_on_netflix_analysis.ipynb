Q.1 > Analyze the amount of content available in different countries ?

```python
# Content available in columns 'Type'

sns.set_style("darkgrid")
ax = sns.countplot(data = df, x='type')
ax.bar_label(ax.containers[0], fontsize=10)
plt.title('The Available Contents', fontsize = 14)
plt.show()
```

Q. 2. How much content of movie and tv show types on Netflix by year?

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df.info()
```

```python
df.head(2)
```

```python
df['date_year'] = df['date_added'].dt.year
df['date_month'] = df['date_added'].dt.month

df.head(2)
```

```python
df['date_year'].value_counts().sort_index().plot(kind='bar')
plt.title('The Number of All Contents on Netflix by Year', fontsize = 14)
plt.xlabel('year', fontsize = 12)
plt.ylabel('count', fontsize = 12)
plt.show()
```

Insight : The amount of content available on netflix has always shown an increase over the years, up to 2000 but reduced in 2020.

Q.3 How to distribute movies and tv shows according to the year of release?

```python
# Separate data for movies and TV shows
df_movie = df[df['type'] == 'Movie']
df_tvshow = df[df['type'] == 'TV Show']

# Count the number of movies and TV shows in each year
count_movie = df_movie['release_year'].value_counts().sort_index()
count_tvshow = df_tvshow['release_year'].value_counts().sort_index()

# Create a figure with two subplots (side by side)
plt.figure(figsize=(15, 5))  # Adjust the figure size

# Create the first subplot for movies
plt.subplot(1, 2, 1)
plt.plot(count_movie.index, count_movie.values , linestyle='-')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movies Released by Year')
plt.grid(True)

# Create the second subplot for TV shows
plt.subplot(1, 2, 2)
plt.plot(count_tvshow.index, count_tvshow.values, linestyle='-')
plt.xlabel('Release Year')
plt.ylabel('Number of TV Shows')
plt.title('Distribution  of TV Shows Released by Year')
plt.grid(True)

# Adjust layout and display both subplots
plt.tight_layout()
plt.show()
```

Insight : In accordance with the graph that shows that more movies are produced, even experiencing an increase from 2010-2019 with a peak point reaching more than 750. but then 2020 experienced a decline in production.

Q.4 Who is the most popular director on Netflix?

The 5 most popular directors in the dataset, according to the content that appears on Netflix.

```python
df_director = df['director'].str.split(', ').explode().value_counts()[1:6]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = df_director.index, y =  df_director.values)
plt.title(' Top 5 Most Popular Director', fontsize =  15)
plt.xlabel('Director', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
plt.show()
```
